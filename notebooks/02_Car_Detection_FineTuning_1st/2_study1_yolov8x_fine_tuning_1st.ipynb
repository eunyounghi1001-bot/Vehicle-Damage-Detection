```python
# [Cell 2] ë°ì´í„°ì…‹ êµ¬ì¶• ë° ë¶„í• 
import os
import shutil
import json
import random
from ultralytics import YOLO
from tqdm import tqdm

# =========================================================
# [ì„¤ì •] ê²½ë¡œ
# =========================================================
SOURCE_ROOT = "/content/drive/MyDrive/03. HDMF/(share)2026_ImageDetectionStudy_No_resizing"
DATASET_ROOT = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification"

# ì˜¤í†  ë¼ë²¨ë§ìš© ëª¨ë¸
auto_label_model = YOLO('yolov8x.pt')
TARGET_CLASSES = [2, 5, 7] # Car, Bus, Truck

# ---------------------------------------------------------
# [í•¨ìˆ˜] bbox ë³€í™˜ (JSON -> YOLO Normalization)
# ---------------------------------------------------------
def convert_bbox_to_yolo(size, box):
    dw = 1. / size[0]
    dh = 1. / size[1]
    x_center = (box[0] + box[2] / 2.0) * dw
    y_center = (box[1] + box[3] / 2.0) * dh
    w = box[2] * dw
    h = box[3] * dh
    return (x_center, y_center, w, h)

# ---------------------------------------------------------
# [í•¨ìˆ˜] ë°ì´í„°ì…‹ êµ¬ì¶• ë©”ì¸ ë¡œì§
# ---------------------------------------------------------
def build_dataset_with_split():
    # ë°ì´í„°ì…‹ í´ë”ê°€ ì´ë¯¸ ìˆìœ¼ë©´ ì‚­ì œí•˜ê³  ë‹¤ì‹œ ë§Œë“¤ì§€ ì—¬ë¶€ í™•ì¸ (ì•ˆì „í•˜ê²Œ ìƒëµ ê°€ëŠ¥)
    # if os.path.exists(DATASET_ROOT): shutil.rmtree(DATASET_ROOT)

    # 1. ì €ì¥ë  í´ë” êµ¬ì¡° ìƒì„±
    for split in ['train', 'val', 'test']:
        os.makedirs(os.path.join(DATASET_ROOT, 'images', split), exist_ok=True)
        os.makedirs(os.path.join(DATASET_ROOT, 'labels', split), exist_ok=True)

    print("ğŸš€ ë°ì´í„°ì…‹ êµ¬ì¶• ì‹œì‘ (Target Folders Only)...")

    # [ìˆ˜ì •ë¨] ì²˜ë¦¬í•  í´ë”ë¥¼ ëª…ì‹œì ìœ¼ë¡œ ì§€ì • (White-list ë°©ì‹)
    target_strategies = {
        "damaged": "json",       # JSON ë¼ë²¨ ì‚¬ìš©
        "normal": "auto_label",  # ì˜¤í†  ë¼ë²¨ë§
        "background": "empty"    # ë¹ˆ íŒŒì¼ (Negative Sample)
    }

    total_processed = 0

    # target_strategiesì˜ í‚¤(folder_name)ë§Œ ìˆœíšŒí•©ë‹ˆë‹¤.
    for folder_name, strategy in target_strategies.items():
        # ì‹¤ì œ ê²½ë¡œ êµ¬ì„±
        img_dir = os.path.join(SOURCE_ROOT, folder_name, "images")
        label_dir = os.path.join(SOURCE_ROOT, folder_name, "labels") # damagedë§Œ ì¡´ì¬

        # í•´ë‹¹ í´ë”ê°€ ì‹¤ì œë¡œ ìˆëŠ”ì§€ í™•ì¸
        if not os.path.exists(img_dir):
            print(f"âš ï¸ ê²½ê³ : '{folder_name}' í´ë”ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê±´ë„ˆëœë‹ˆë‹¤.")
            continue

        # ì´ë¯¸ì§€ íŒŒì¼ë§Œ ë¦¬ìŠ¤íŠ¸ì—…
        files = [f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

        if not files:
            print(f"âš ï¸ ê²½ê³ : '{folder_name}' í´ë”ì— ì´ë¯¸ì§€ê°€ ì—†ìŠµë‹ˆë‹¤.")
            continue

        random.shuffle(files) # ëœë¤ ì„ê¸°

        print(f"Processing [{folder_name}]: {len(files)}ì¥ (Strategy: {strategy})")

        # ë¶„í•  ì§€ì  ê³„ì‚° (7:2:1)
        n_files = len(files)
        n_train = int(n_files * 0.7)
        n_val = int(n_files * 0.9)

        for i, filename in enumerate(tqdm(files, desc=folder_name)):
            # Split ê²°ì •
            if i < n_train: split = 'train'
            elif i < n_val: split = 'val'
            else: split = 'test'

            src_img = os.path.join(img_dir, filename)
            dst_img = os.path.join(DATASET_ROOT, 'images', split, filename)
            dst_txt = os.path.join(DATASET_ROOT, 'labels', split, os.path.splitext(filename)[0] + ".txt")

            # A. ì´ë¯¸ì§€ ë³µì‚¬
            shutil.copy(src_img, dst_img)

            # B. ë¼ë²¨ ìƒì„± ë¡œì§
            if strategy == "json":
                json_file = os.path.splitext(filename)[0] + ".json"
                json_path = os.path.join(label_dir, json_file)

                # JSON íŒŒì¼ì´ ìˆëŠ” ê²½ìš°ë§Œ ì²˜ë¦¬
                if os.path.exists(json_path):
                    with open(json_path, 'r') as f:
                        data = json.load(f)
                        w_img, h_img = data['images']['width'], data['images']['height']

                        with open(dst_txt, 'w') as out:
                            for ann in data['annotations']:
                                bbox = ann['bbox']
                                yolo_box = convert_bbox_to_yolo((w_img, h_img), bbox)
                                # class_id 0 (Vehicle)ìœ¼ë¡œ í†µì¼
                                out.write(f"0 {yolo_box[0]:.6f} {yolo_box[1]:.6f} {yolo_box[2]:.6f} {yolo_box[3]:.6f}\n")

            elif strategy == "auto_label":
                # ê¸°ì¡´ ëª¨ë¸ ì¶”ë¡  (conf ë‚®ê²Œ ì„¤ì •í•  í•„ìš” ì—†ìŒ, Normal ì°¨ëŸ‰ì€ ì˜ ì°¾ìœ¼ë¯€ë¡œ ê¸°ë³¸ê°’ ì‚¬ìš©)
                results = auto_label_model(src_img, conf=0.25, verbose=False)
                with open(dst_txt, 'w') as out:
                    for result in results:
                        for box in result.boxes:
                            if int(box.cls[0]) in TARGET_CLASSES:
                                x, y, w, h = box.xywhn[0].tolist()
                                out.write(f"0 {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")

            elif strategy == "empty":
                # ë¹ˆ txt íŒŒì¼ ìƒì„± (Explicit Negative)
                with open(dst_txt, 'w') as out:
                    pass

            total_processed += 1

    # 2. data.yaml ìƒì„± (Testì…‹ ì œì™¸ - í•™ìŠµìš©)
    yaml_content = f"""
path: {DATASET_ROOT}
train: images/train
val: images/val
names:
  0: Vehicle
"""
    yaml_path = os.path.join(DATASET_ROOT, 'yolov8_tuning_251207.yaml')
    with open(yaml_path, 'w') as f:
        f.write(yaml_content)

    print(f"\nâœ… ë°ì´í„°ì…‹ êµ¬ì¶• ì™„ë£Œ! (ì´ {total_processed}ì¥)")
    print(f"   ì„¤ì • íŒŒì¼: {yaml_path}")

# ì‹¤í–‰
build_dataset_with_split()
```

```python
# [Cell] í˜„ì¬ ë°ì´í„°ì…‹ ë¶„í•  í˜„í™© í™•ì¸
import os

# 1. í™•ì¸í•  ë°ì´í„°ì…‹ ë£¨íŠ¸ ê²½ë¡œ
DATASET_ROOT = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification"

def check_dataset_status():
    if not os.path.exists(DATASET_ROOT):
        print(f"âŒ ë°ì´í„°ì…‹ í´ë”ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {DATASET_ROOT}")
        return

    splits = ['train', 'val', 'test']
    counts = {}

    # ê° í´ë”ë³„ ì´ë¯¸ì§€ ê°œìˆ˜ ì¹´ìš´íŠ¸
    for split in splits:
        dir_path = os.path.join(DATASET_ROOT, 'images', split)
        if os.path.exists(dir_path):
            # ì´ë¯¸ì§€ íŒŒì¼ë§Œ ì¹´ìš´íŠ¸ (.jpg, .png ë“±)
            files = [f for f in os.listdir(dir_path) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
            counts[split] = len(files)
        else:
            counts[split] = 0

    # ì „ì²´ í•©ê³„
    total = sum(counts.values())

    # ê²°ê³¼ ì¶œë ¥
    print(f"\n--- ìµœì¢… ë¶„ë¦¬ ê²°ê³¼ (ë°ì´í„°ì…‹ ê²½ë¡œ: {os.path.basename(DATASET_ROOT)}) ---")
    print(f"ì „ì²´ ìœ íš¨ ë°ì´í„° ìˆ˜: {total}")

    if total > 0:
        # Train
        ratio_train = counts['train'] / total
        print(f"âœ… Train ë°ì´í„° ìˆ˜: {counts['train']} ({ratio_train:.2f})")
        # Val
        ratio_val = counts['val'] / total
        print(f"âœ… Valid ë°ì´í„° ìˆ˜: {counts['val']} ({ratio_val:.2f})")
        # Test
        ratio_test = counts['test'] / total
        print(f"âœ… Test  ë°ì´í„° ìˆ˜: {counts['test']} ({ratio_test:.2f})")
    else:
        print("âš ï¸ ë°ì´í„°ê°€ í•˜ë‚˜ë„ ì—†ìŠµë‹ˆë‹¤. ë°ì´í„°ì…‹ êµ¬ì¶• ì½”ë“œë¥¼ ë¨¼ì € ì‹¤í–‰í•´ì£¼ì„¸ìš”.")

# ì‹¤í–‰
check_dataset_status()
```

```python
# [Cell 3] Fine-tuning ì‹œì‘
import os
import shutil
import json
import random
from ultralytics import YOLO
from tqdm import tqdm

# =========================================================
# [ì„¤ì •] ê²½ë¡œ
# =========================================================
SOURCE_ROOT = "/content/drive/MyDrive/03. HDMF/(share)2026_ImageDetectionStudy_No_resizing"
DATASET_ROOT = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification"

# =========================================================
# [ì„¤ì •] íŒŒë¼ë¯¸í„°
# =========================================================
YAML_PATH = os.path.join(DATASET_ROOT, 'yolov8_tuning_251207.yaml')
MODEL_NAME = 'yolov8x.pt' # Pre-trained ëª¨ë¸
EPOCHS = 50               # í•™ìŠµ íšŸìˆ˜ (15~20 ì¶”ì²œ)
BATCH_SIZE = 16           # ë°°ì¹˜ í¬ê¸° (OOM ë°œìƒ ì‹œ 8ë¡œ ì¤„ì„)
IMG_SIZE = 640            # ì´ë¯¸ì§€ í¬ê¸°

def run_training():
    print(f"ğŸ”¥ í•™ìŠµ ì‹œì‘: {MODEL_NAME} -> Epochs: {EPOCHS}")
    SAVE_DIR = os.path.join(DATASET_ROOT, "train_results")

    # ëª¨ë¸ ë¡œë“œ
    model = YOLO(MODEL_NAME)

    # í•™ìŠµ ì‹¤í–‰
    model.train(
        data=YAML_PATH,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        batch=BATCH_SIZE,
        project=SAVE_DIR,
        name='yolov8x_finetuned_vehicle', # ê²°ê³¼ ì €ì¥ í´ë”ëª… (runs/detect/...)
        exist_ok=True,
        optimizer='AdamW',
        lr0=1e-4,         # Fine-tuningìš© ë‚®ì€ í•™ìŠµë¥ 
        freeze=10,        # ì•ë‹¨ ë ˆì´ì–´ ê³ ì •
        patience=15,       # ì¡°ê¸° ì¢…ë£Œ ì„¤ì •
        verbose=True
    )
    print("ğŸ‰ í•™ìŠµ ì™„ë£Œ!")

# ì‹¤í–‰
run_training()
```

# 4. test ë°ì´í„° ì ìš© ê²°ê³¼ ë¶„ì„

```python
# [Cell] Fine-tuned Model Test Set í‰ê°€
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from tqdm import tqdm

# =========================================================
# [ì„¤ì •] ê²½ë¡œ ë° íŒŒë¼ë¯¸í„°
# =========================================================
# 1. í•™ìŠµëœ ëª¨ë¸ ê²½ë¡œ (User í™˜ê²½ì— ë§ê²Œ ìˆ˜ì • í•„ìš”)
#    (ë°©ê¸ˆ í•™ìŠµí•œ runs í´ë” ë˜ëŠ” ë“œë¼ì´ë¸Œì— ì €ì¥ëœ ê²½ë¡œ)
MODEL_PATH = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/train_results/yolov8x_finetuned_vehicle/weights/best.pt"

# 2. í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ ê²½ë¡œ
TEST_IMG_DIR = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/images/test"
TEST_LABEL_DIR = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/labels/test"

# 3. ê²°ê³¼ ì €ì¥ ê²½ë¡œ
RESULT_ROOT = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification"
os.makedirs(RESULT_ROOT, exist_ok=True)

CONF_THRESHOLD = 0.25  # Fine-tuning í–ˆìœ¼ë¯€ë¡œ í‘œì¤€ ì„ê³„ê°’ ì‚¬ìš©

def evaluate_finetuned_model():
    if not os.path.exists(MODEL_PATH):
        print(f"âŒ ëª¨ë¸ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤: {MODEL_PATH}")
        return

    print(f"ğŸš€ í‰ê°€ ì‹œì‘...")
    print(f"   - ëª¨ë¸: {os.path.basename(MODEL_PATH)}")
    print(f"   - ë°ì´í„°: {TEST_IMG_DIR}")

    model = YOLO(MODEL_PATH)

    # í‰ê°€ ëŒ€ìƒ íŒŒì¼ ë¦¬ìŠ¤íŠ¸ì—…
    image_files = [f for f in os.listdir(TEST_IMG_DIR) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

    results_list = []
    speed_stats = []

    print(f"   - ì´ í‰ê°€ ëŒ€ìƒ: {len(image_files)}ì¥")

    for file in tqdm(image_files, desc="Inference"):
        img_path = os.path.join(TEST_IMG_DIR, file)
        label_path = os.path.join(TEST_LABEL_DIR, os.path.splitext(file)[0] + ".txt")

        # -----------------------------------------------------
        # 1. ì •ë‹µ(True Label) í™•ì¸ ë¡œì§
        # -----------------------------------------------------
        # ë¼ë²¨ íŒŒì¼ì´ ìˆê³ , ë‚´ìš©ì´ ë¹„ì–´ìˆì§€ ì•Šìœ¼ë©´ Vehicle(1)
        true_label = 0
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                content = f.read().strip()
                if len(content) > 0:
                    true_label = 1

        # -----------------------------------------------------
        # 2. ëª¨ë¸ ì¶”ë¡  (Prediction)
        # -----------------------------------------------------
        results = model(img_path, conf=CONF_THRESHOLD, verbose=False)

        # ì†ë„ ì¸¡ì •
        speed_info = results[0].speed
        total_time_ms = speed_info['preprocess'] + speed_info['inference'] + speed_info['postprocess']
        speed_stats.append(total_time_ms)

        # ì˜ˆì¸¡ ë¼ë²¨ ê²°ì • (ë°•ìŠ¤ê°€ í•˜ë‚˜ë¼ë„ ìˆìœ¼ë©´ Vehicle)
        pred_label = 0
        for r in results:
            if len(r.boxes) > 0:
                pred_label = 1
                break

        # ê²°ê³¼ ì €ì¥
        results_list.append({
            "filename": file,
            "true_label": true_label,
            "pred_label": pred_label,
            "is_correct": (true_label == pred_label),
            "full_path": img_path
        })

    # ë°ì´í„°í”„ë ˆì„ ë³€í™˜
    df = pd.DataFrame(results_list)

    # =========================================================
    # [ë¶„ì„ ê²°ê³¼ ë¦¬í¬íŠ¸] (ìš”ì²­í•˜ì‹  ì½”ë“œ í¬ë§· ì ìš©)
    # =========================================================
    print("\n" + "="*50)
    print("ğŸ“Š [ìµœì¢… ë¶„ì„ ê²°ê³¼ - Fine-tuned Model]")
    print("="*50)

    # A. ì •í™•ë„ í‰ê°€
    acc = accuracy_score(df['true_label'], df['pred_label'])
    print(f"âœ… 1. ì •í™•ë„ (Accuracy): {acc:.4f} ({acc*100:.2f}%)")

    # B. ì†ë„ í‰ê°€
    if speed_stats:
        avg_time = np.mean(speed_stats)
        min_time = np.min(speed_stats)
        max_time = np.max(speed_stats)
        fps = 1000 / avg_time

        print(f"\nâš¡ 2. ì¶”ë¡  ì†ë„ (Inference Speed):")
        print(f"   - í‰ê·  ì†Œìš” ì‹œê°„ : {avg_time:.2f} ms/ì¥")
        print(f"   - ìµœì†Œ ì†Œìš” ì‹œê°„ : {min_time:.2f} ms")
        print(f"   - ìµœëŒ€ ì†Œìš” ì‹œê°„ : {max_time:.2f} ms")
        print(f"   - ì²˜ë¦¬ëŸ‰ (FPS)   : {fps:.2f} FPS")

    # C. ìƒì„¸ ë¶„ë¥˜ ë¦¬í¬íŠ¸
    print("\nğŸ“ 3. ìƒì„¸ ë¶„ë¥˜ ë¦¬í¬íŠ¸:")
    print(classification_report(df['true_label'], df['pred_label'], target_names=['Non-Vehicle', 'Vehicle']))

    # D. í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
    plt.figure(figsize=(6, 5))
    cm = confusion_matrix(df['true_label'], df['pred_label'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Pred: Non-Vehicle', 'Pred: Vehicle'],
                yticklabels=['True: Non-Vehicle', 'True: Vehicle'])
    plt.title(f'Confusion Matrix\n(Fine-tuned Result)')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

    # =========================================================
    # E. ê²°ê³¼ ì €ì¥
    # =========================================================
    save_filename = f"inference_finetuned_test_results.csv"
    save_path = os.path.join(RESULT_ROOT, save_filename)

    df.to_csv(save_path, index=False, encoding='utf-8-sig')

    print(f"\nğŸ’¾ ì „ì²´ ì‹¤í–‰ ê²°ê³¼ ì €ì¥ ì™„ë£Œ!")
    print(f"   -> íŒŒì¼ ê²½ë¡œ: {save_path}")
    print(f"   -> ì´ ë°ì´í„° ìˆ˜: {len(df)}ê±´")

    failed_count = len(df[df['is_correct'] == False])
    print(f"   -> (ì„±ê³µ: {len(df)-failed_count}ê±´, ì‹¤íŒ¨: {failed_count}ê±´)")

# ì‹¤í–‰
evaluate_finetuned_model()
```

# 5. ì›ì¸ ì¶”ì •
 - ì›ì¸ ì¶”ì •: "ë¼ë²¨ì˜ ë¶ˆì¼ì¹˜(Label Conflict)"
    - ê°€ì¥ ìœ ë ¥í•œ ì›ì¸ì€ í•™ìŠµ ë°ì´í„° ê°„ì˜ "ì •ë‹µ ê¸°ì¤€"ì´ ì„œë¡œ ë‹¤ë¥´ê¸° ë•Œë¬¸ì¼ ê°€ëŠ¥ì„±

 - Damaged ë°ì´í„° (JSON ê¸°ë°˜):
    - JSONì— ìˆë˜ bboxê°€ **"ì°¨ëŸ‰ ì „ì²´"**ê°€ ì•„ë‹ˆë¼ **"íŒŒì†ëœ ë¶€ìœ„(ìŠ¤í¬ë˜ì¹˜, ì°Œê·¸ëŸ¬ì§ ë“± ì‘ì€ ì˜ì—­)"**ë§Œ ê°ì‹¸ê³  ìˆì—ˆì„ í™•ë¥ ì´ ë†’ìŒ.
 - Normal ë°ì´í„° (Auto-labeling):
    - YOLOê°€ ìë™ìœ¼ë¡œ ë¼ë²¨ë§í–ˆìœ¼ë¯€ë¡œ **"ì°¨ëŸ‰ ì „ì²´"**ë¥¼ ì¡ì•˜ìŒ

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import cv2
import os
import random
from ultralytics import YOLO

# =========================================================
# [ì„¤ì •] ê²½ë¡œ
# =========================================================
# 1. í•™ìŠµëœ ìµœì  ëª¨ë¸ ê²½ë¡œ (í™•ì¸ í•„ìš”)
BEST_MODEL_PATH = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/train_results/yolov8x_finetuned_vehicle/weights/best.pt"

# 2. í…ŒìŠ¤íŠ¸ ë°ì´í„° ê²½ë¡œ
TEST_IMG_DIR = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/images/test"
TEST_LABEL_DIR = "/content/drive/MyDrive/03. HDMF/(pre_study)2026_ImageDetectionStudy/1_car_classification/labels/test"

def visualize_comparison(num_samples=5):
    if not os.path.exists(BEST_MODEL_PATH):
        print("âŒ í•™ìŠµëœ ëª¨ë¸ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
        return

    model = YOLO(BEST_MODEL_PATH)
    files = [f for f in os.listdir(TEST_IMG_DIR) if f.lower().endswith(('.jpg', '.png'))]
    random.shuffle(files)

    samples = files[:num_samples]

    plt.figure(figsize=(15, 5 * num_samples))

    for i, file in enumerate(samples):
        img_path = os.path.join(TEST_IMG_DIR, file)
        label_path = os.path.join(TEST_LABEL_DIR, os.path.splitext(file)[0] + ".txt")

        # ì´ë¯¸ì§€ ë¡œë“œ
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        h, w, _ = img.shape

        # -------------------------------------------------
        # 1. Ground Truth (ì´ˆë¡ìƒ‰) ê·¸ë¦¬ê¸°
        # -------------------------------------------------
        if os.path.exists(label_path):
            with open(label_path, 'r') as f:
                lines = f.readlines()
                for line in lines:
                    cls, x, y, bw, bh = map(float, line.strip().split())
                    # YOLO xywhn -> pixel xymn (ì¢Œìƒë‹¨)
                    x1 = int((x - bw / 2) * w)
                    y1 = int((y - bh / 2) * h)
                    x2 = int((x + bw / 2) * w)
                    y2 = int((y + bh / 2) * h)

                    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 3) # Green
                    cv2.putText(img, "GT", (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        # -------------------------------------------------
        # 2. Prediction (ë¹¨ê°„ìƒ‰) ê·¸ë¦¬ê¸°
        # -------------------------------------------------
        results = model(img_path, conf=0.1, verbose=False) # Threshold ë‚®ê²Œ ì„¤ì •
        for result in results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                conf = float(box.conf[0])

                cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 3) # Red
                cv2.putText(img, f"Pred {conf:.2f}", (x1, y2+30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

        # ì¶œë ¥
        plt.subplot(num_samples, 1, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Sample: {file} (Green: GT, Red: Prediction)", fontsize=14)

    plt.tight_layout()
    plt.show()
```
